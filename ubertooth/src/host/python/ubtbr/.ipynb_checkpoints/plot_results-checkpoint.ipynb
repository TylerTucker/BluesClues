```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
def read_file(filename):
    f = open(filename, 'r')
    xs = []
    ys = []
    count = 1
    for x in f:
        xs.append(count)
        ys.append(float(x.strip()))
        count += 1
    return xs, ys
```

```python
xs, stingray = read_file('results.txt')
# xs, straight = read_file('gainesville_03_15/straight_results.txt')
```

```python
plt.scatter(xs, stingray, c='orange', label='Stingray')
# plt.scatter(xs, straight, c='b', label='Direct')
plt.xlabel('Sample Number')
plt.ylabel('Delay (s)')
plt.title('')
plt.legend()
plt.show()
```
